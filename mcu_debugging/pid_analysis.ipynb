```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_file_dir = "./pid_data/"
```

```python
'''
original setting
float Kp_one = 1.15;             // proportional
float Ki_one = 2.235;            // integral
float Kd_one = 0.00025;          // differential
'''

data_csv = pd.read_csv(data_file_dir + 'pid_data.csv', header=None)
data_csv.drop(columns=[0])

data_column1_parse = np.array([np.float32(i[13:]) for i in data_csv[1]])
data_column2_parse = np.array([np.float32(i[11:]) for i in data_csv[2]])
plt.ylabel('speed (RPM)')
plt.xlabel('samples')
plt.plot(data_column1_parse, label="actual")
plt.plot(data_column2_parse, label="target")
plt.legend()
plt.show()
```

```python
# Finalized version of PID control

# right motor
data_csv_pd_only_2 = pd.read_csv(data_file_dir + 'pid_temp_R.csv', header=None)
data_csv_pd_only_2.drop(columns=[0])
data_csv_pd_only_col2 = np.array([np.float32(i[13:]) for i in data_csv_pd_only_2[1][:200]])
data_csv_pd_only_col3 = np.array([np.float32(i[11:]) for i in data_csv_pd_only_2[2][:200]])
plt.plot(data_csv_pd_only_col3, 'green', label='target')
plt.plot(data_csv_pd_only_col2, label='right actual')

# left motor 
l_data_csv_pd_only_2 = pd.read_csv(data_file_dir + 'pid_temp_L.csv', header=None)
l_data_csv_pd_only_2.drop(columns=[0])
l_data_csv_pd_only_col2 = np.array([np.float32(i[13:]) for i in l_data_csv_pd_only_2[1][:200]])
l_data_csv_pd_only_col3 = np.array([np.float32(i[11:]) for i in l_data_csv_pd_only_2[2][:200]])
# plt.plot(l_data_csv_pd_only_col3, 'green', label='target')
plt.plot(l_data_csv_pd_only_col2, label='left actual')
# plt.plot(l_data_csv_pd_only_col2, label='actual')
plt.title("PID parameter for both motors")
plt.ylabel('speed (RPM)')
plt.xlabel('samples')
plt.legend()
plt.show()
```
